## Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)#with same col name

for df in pd.read_csv('modified.csv', chunksize=5):#5 row df
    #print("Chunk Dataframe: ")
    #print(df)
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
    print(new_df)
```
